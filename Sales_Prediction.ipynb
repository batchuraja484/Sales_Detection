```python
X=np.array(df_final).astype('float32')
Y=np.array(df_target).astype('float32')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.15)
x_test,x_val,y_test,y_val=train_test_split(x_test,y_test,test_size=0.5)
```

```python
model=xg.XGBRegressor(objective='reg:squarederror',learning_rate=0.1,max_depth=5,n_estimators=100)
```

```python
model.fit(x_train,y_train)
```

```python
result=model.score(x_test,y_test)
print("Accuracy Score of the Test set is {}".format(result))
```

```python
print("Accuray Score for the Trian data ",model.score(x_train,y_train))
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error,adjusted_rand_score

r2_scr=r2_score(y_test,y_predict)
MAE_Err=mean_absolute_error(y_test,y_predict)
MSE_Err=mean_absolute_error(y_test,y_predict)
RMSE=format(np.sqrt(MSE_Err),'.3f')

print("R2_Score=",r2_scr,"\nMAE_ERROR=",MAE_Err,"\nMSE_ERROR=",MSE_Err,"\nRMSE ERROR=",RMSE)
```

```python
#adjusted R2 Score
#formula is 1-(1-r2)*(n-1)/(n-k-1)
#where n is no of test points
#k is no input vairables for the prediction x_test.shape[1]

n=len(x_test)
k=x_test.shape[1]
adj_r2=1-(1-r2_scr)*(n-1)/(n-k-1)
print("Adjusted R2 Score is:",adj_r2)
```
