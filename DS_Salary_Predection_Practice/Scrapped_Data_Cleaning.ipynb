```python
import pandas as pd
import numpy as np

```

```python
df=pd.read_csv('glassdoor_jobs_scraped.csv')
```

```python
df.head(1)
```

```python
#Need to check the salary per hour
df[df['Salary Estimate'].str.lower().str.contains('hour')].head(1)
```

```python
#Parsing Salary
df['hour']=df['Salary Estimate'].apply(lambda x: 1 if 'hour' in x.lower() else 0)
```

```python
df.query('hour==1').head(1)
```

```python
df[df['hour']==1].head(1)
```

```python
df.where(df['hour']==1).head(1)
```

```python
df=df[df['Salary Estimate']!='-1']
```

```python
len(df)
```

```python
print(df['Salary Estimate'].apply(lambda x: x.split('(')[0]).head(1))
```

```python
df['salary']=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df['salary']=df['salary'].apply(lambda x:x.replace('$','').replace('K','').replace('Employer Provided Salary:','').replace('Per Hour',''))
```

```python
df['min_salray']=df['salary'].apply(lambda x:x.split('-')[0])
df['max_salary']=df['salary'].apply(lambda x:x.split('-')[1])
```

```python
df['average_salary']=(df.min_salray.astype(int)+df.max_salary.astype(int))/2
```

```python
df.head(5)
```

```python
#Need to create the company text,
df['Company_txt']=df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
```

```python
df.head(1)
```

```python
#state field 
df['State']=df.apply(lambda x: 1 if x.Location==x.Headquarters else 0, axis=1)
```

```python
df.head(1)
```

```python
#age of company
df['age']=df.Founded.apply(lambda x: x if x<0 else 2020-x)
```

```python
df.head(1)
```

```python
#parsing of job description (python, etc.)
```

```python
#python
df['Python']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df.head(1)
```

```python
#r studio 
df['R Studio']=df['Job Description'].apply(lambda x:1 if 'r studio' in x.lower() else 0)
df.head(1)
```

```python
#spark 
df['spark']=df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
```

```python
#aws
df['aws']=df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
#excel
df['excel']=df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```
